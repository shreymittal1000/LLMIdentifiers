```python
from collections import defaultdict
import json
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import os
import pandas as pd
import seaborn as sns
from sklearn.metrics import precision_score, recall_score, confusion_matrix
```

```python
# Load data
folder_path = "../results/base/"
data = []

for root, dirs, files in os.walk(folder_path):
    for filename in files:
        if filename.endswith(".json"):
            file_path = os.path.join(root, filename)
            try:
                with open(file_path, "r") as f:
                    entry = json.load(f)
                    data.append(entry)
            except json.JSONDecodeError as e:
                print(f"Skipped invalid JSON: {file_path} ({e})")

print(f"Loaded {len(data)} valid JSON result files from '{folder_path}' and its subfolders.")
```

```python
from sklearn.metrics import confusion_matrix
import pandas as pd

# make sure y_true and y_pred are your flat lists/arrays of lowercase labels
labels = sorted(["gpt", "llama", "gemini", "claude", 
                 "qwen", "deepseek", "mistral"])

y_true, y_pred = [], []
for entry in data:
    y_true.append(entry['model_general_name_1'])  # agent_0 guess target
    y_pred.append(entry['agent_0_guess'])
    y_true.append(entry['model_general_name_0'])  # agent_1 guess target
    y_pred.append(entry['agent_1_guess'])

y_true = np.array(y_true)
y_pred = np.array(y_pred)


# 1) Raw counts confusion matrix
cm_counts = confusion_matrix(y_true, y_pred, labels=labels)
cm_counts_df = pd.DataFrame(cm_counts, index=labels, columns=labels)

# 2) Pairwise recall matrix:
#    each row sums to 1 →  recall_matrix[i,j] = TP_ij / (TP_ij + FN_ij)
cm_recall = confusion_matrix(y_true, y_pred,
                             labels=labels,
                             normalize='true')
recall_df = pd.DataFrame(cm_recall, index=labels, columns=labels)

# 3) Pairwise precision matrix:
#    each column sums to 1 →  precision_matrix[i,j] = TP_ij / (TP_ij + FP_ij)
cm_precision = confusion_matrix(y_true, y_pred,
                                labels=labels,
                                normalize='pred')
precision_df = pd.DataFrame(cm_precision, index=labels, columns=labels)

# 4) (Optional) Map to pretty names
recall_df = recall_df.rename(index=pretty_names, columns=pretty_names)
precision_df = precision_df.rename(index=pretty_names, columns=pretty_names)

# 5) Display
print("Raw counts:\n", cm_counts_df, "\n")
print("Recall (true→pred):\n", recall_df, "\n")
print("Precision (true→pred):\n", precision_df)
```

```python
# 4) Plotting helper
def plot_heatmap(df, title, save_file_name):
    plt.figure(figsize=(14, 12))
    ax = sns.heatmap(
        df,
        annot=True, fmt='.2f',
        cmap='YlGnBu',
        linewidths=0.3,
        cbar_kws={"shrink": 0.7},
        annot_kws={"fontsize": 24}
    )
    ax.set_title(title, fontsize=30)
    
    xticks = ax.get_xticklabels()
    yticks = ax.get_yticklabels()
    x_labels = [pretty_names.get(label.get_text(), label.get_text()) for label in xticks]
    y_labels = [pretty_names.get(label.get_text(), label.get_text()) for label in yticks]
    ax.set_xticklabels(x_labels, rotation=45, ha='right', fontsize=28)
    ax.set_yticklabels(y_labels, rotation=0, fontsize=28)
    
    ax.set_xlabel('Guessed Model', fontsize=28)
    ax.set_ylabel('Actual Model', fontsize=28)
    plt.xticks(rotation=45, ha='right')
    plt.yticks(rotation=0)
    cbar = ax.collections[0].colorbar
    cbar.ax.tick_params(labelsize=28)

    plt.tight_layout()
    plt.savefig("figures_and_files/" + save_file_name, dpi=300, bbox_inches='tight')
    plt.show()

# 5) Display both metrics
plot_heatmap(precision_df, 'Precision by Model', 'precision.eps')
plot_heatmap(recall_df, 'Recall by Model', 'recall.eps')
```
